## olist order dataset

```python
# Display info and head of orders_data
print(orders_data.info())
orders_data.head(2)
```

```python
# converting date columns to datetime format
date_col = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date',
            'order_delivered_customer_date','order_estimated_delivery_date']

for column in date_col:
  orders_data[column] = pd.to_datetime(orders_data[column])
```

```python
orders_data.info()
```

```python
# Analyze the distribution of order_status
orders_data['order_status'].value_counts()
```

```python
# filtering only delivered orders because 
# other orders are not completed yet then only delivered orders for revenue analysis
orders_data = orders_data[orders_data['order_status']=='delivered']
orders_data.shape
```

```python
orders_data.isnull().sum()
```

In orders data order_approved_at , order_delivered_carrier_date, order_delivered_customer_date column are having Null values which means order is not approved or never shipped or orer is not delivered yet/ cancelled 
therefore, Drop the null values rows 

```python
# Drop rows with missing 'order_delivered_customer_date' order_delivered_carrier_date', 'order_approved_at'
orders_delivered = orders_data.dropna(subset=['order_delivered_customer_date', 
                                              'order_delivered_carrier_date', 'order_approved_at'])
```
